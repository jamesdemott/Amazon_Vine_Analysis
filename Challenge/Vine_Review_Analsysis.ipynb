```python
import pandas as pd
```

```python
df = pd.read_csv("vine_analysis_data_with_headings.csv")
df.dtypes
```

```python
df_relevant = df.loc[df['total_votes'] > 20]
```

```python
df['helpful_div_total'] = df['helpful_votes'] / df['total_votes']
```

```python
df_helpful = df.loc[df['helpful_div_total']> .5]
df_helpful
```

```python
df_helpful_dropped = df_helpful.drop(['helpful_div_total'], axis = 1)
df_helpful_dropped
```

```python
df_paid = df_helpful_dropped.loc[df_helpful_dropped['vine'] == 'Y']
```

```python
df_paid
```

```python
df_unpaid = df_helpful_dropped.loc[df_helpful_dropped['vine'] == 'N']
```

```python
df_unpaid
```

```python
## Analysis for Paid 
# total number of reviews
# the number of 5-star reviews 
# and the percentage of 5-star reviews as a percentage of total
```

```python
paid_count = df_paid.count()
paid_count = paid_count["review_id"]
```

```python
unpaid_count = df_unpaid.count()
unpaid_count = unpaid_count["review_id"]
```

```python
total_5_star_paid = df_paid.loc[df_paid["star_rating"]== 5]
total_5_star_paid = total_5_star_paid.count()
total_5_star_paid = total_5_star_paid["review_id"]
```

```python
total_5_star_unpaid = df_unpaid.loc[df_unpaid["star_rating"]==5]
total_5_star_unpaid = total_5_star_unpaid.count()
total_5_star_unpaid = total_5_star_unpaid["review_id"]
```

```python
paid_percentage_5 =  total_5_star_paid / paid_count
paid_percentage_5
```

```python
unpaid_percentage_5 = total_5_star_unpaid / unpaid_count
```

```python
#Analysis Summary

print ("______________________Total Review Counts______________________")
print(f"The total number of paid reviews was "+"{:,}".format(paid_count))
print(f"The total number of unpaid reviews was "+"{:,}".format(unpaid_count))
print("____________________Total Five Star Reviews____________________")
print (f"The total number of 5 star review for paid products was "+"{:,}".format(total_5_star_paid))
print (f"The total number of 5 star review for unpaid products was "+"{:,}".format(total_5_star_unpaid))
print("____________________Ratio of 5 Star Reviews_____________________")
print(f"For paid product, the total percentage of 5 star reviews was "+"{:.1%}".format(paid_percentage_5))
print(f"For unpaid product, the total percentage of 5 star reviews was "+"{:.1%}".format(unpaid_percentage_5))
```
